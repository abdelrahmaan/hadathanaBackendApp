```python
from firecrawl import Firecrawl

app = Firecrawl(api_key="fc-a0e6b09c69d5441293d77c29a403ae85")
result = app.scrape("https://shamela.ws/book/1681/8808",
    formats=[{
        "type": "json",
        "prompt": """Extract data from this page and return a JSON object with exactly this structure:
{
  "status": "success",
  "book_id": <integer book ID from the URL>,
  "page_number": <integer page number from the URL>,
  "url": "<full page URL>",
  "breadcrumb_links": [{"text": "<link text>", "href": "<full URL>"}],
  "hadith_blocks": [
    {
      "full_text": "<complete hadith text including chain of narrators and matn>",
      "matn": "<the matn text only, typically inside quotes or marked distinctly>",
      "narrators": [{"id": "<narrator ID from /narrator/<id> URL>", "name": "<narrator name>", "url": "<full narrator URL>"}]
    }
  ]
}
Extract all hadith blocks on the page. For each hadith, extract the full text, the matn (hadith body), and all narrators with their IDs and URLs."""
    }],
    only_main_content=False,
    timeout=120000
)
print(result)
```

```python
pprint(result.json)
```

```python
from firecrawl import Firecrawl
from pydantic import BaseModel

app = Firecrawl(api_key="fc-a0e6b09c69d5441293d77c29a403ae85")

class BreadcrumbLink(BaseModel):
    text: str
    href: str

class Narrator(BaseModel):
    id: str
    name: str
    url: str

class HadithBlock(BaseModel):
    full_text: str
    matn: str
    narrators: list[Narrator]

class PageData(BaseModel):
    status: str
    book_id: int
    page_number: int
    url: str
    breadcrumb_links: list[BreadcrumbLink]
    hadith_blocks: list[HadithBlock]

result = app.scrape(
    "https://shamela.ws/book/1681/8808",
    formats=[{
        "type": "json",
        "schema": PageData.model_json_schema()
    }],
    only_main_content=False,
    timeout=120000
)

print(result.json)
```

```python
pprint(result.json)
```
